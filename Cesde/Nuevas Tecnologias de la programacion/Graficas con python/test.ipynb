```python
import pandas as pd
```

```python
file = 'carros.xlsx'
data = pd.read_excel(file)
df = pd.DataFrame(data)
display(df)
```

```python
#Preguntas Generales:

#¿Cuántos registros hay en total en el DataFrame?

total_registros = len(df)
print(f"Total de registros: {total_registros}")
```

```python
#¿Cuántas columnas tiene el DataFrame y cuáles son sus nombres?

columnas = df.columns
cantidad_columnas = len(columnas)
print(f"Cantidad de columnas: {cantidad_columnas}")
print(f"Nombre de las columnas: {list(columnas)}")
```

```python

#Análisis de Datos:

#¿Cuál es la media de precios de venta (selling_price)?

media_precio = df['selling_price'].mean()
print(f"Media de precios de venta: {media_precio}")
```

```python
#¿Cuál es la cantidad total de kilómetros recorridos (km_driven)?

total_kilometros = df['km_driven'].sum()
print(f"Total de kilómetros recorridos: {total_kilometros}")
```

```python
#Agrupación y Resumen:

#¿Cuántos registros hay para cada tipo de combustible (fuel)?

conteo_combustible = df['fuel'].value_counts()
print("Conteo de registros por tipo de combustible:")
print(conteo_combustible)
```

```python
#¿Cuál es la venta total agrupada por año (year)?

venta_total_por_anio = df.groupby('year')['selling_price'].sum()
print("Venta total agrupada por año:")
print(venta_total_por_anio)
```

```python
#Filtrado de Datos:

#¿Cuántos registros hay para propietarios de segunda mano (owner = 'Second Owner')?

propietarios_segunda_mano = df[df['owner'] == 'Second Owner']
cantidad_segunda_mano = len(propietarios_segunda_mano)
print(f"Registros con propietarios de segunda mano: {cantidad_segunda_mano}")
```

```python
#¿Cuántos registros hay para transmisiones automáticas (transmission = 'Automatic')?

transmision_automatica = df[df['transmission'] == 'Automatic']
cantidad_automatica = len(transmision_automatica)
print(f"Registros con transmisión automática: {cantidad_automatica}")
```

```python
df.to_json('df_resultante.json', orient='records')
```
