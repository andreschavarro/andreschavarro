```python

# primero importamos la libreria de matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# como primer argumento del metodo se espera un array 
# ya como segundo argumento le podemos poner ro que significa r de red y o de points y pongo una s me salen cuadrados
# la primera letra siempre sera el color como g de green y la segunda letra sera del tipo de forma
```

```python

# Datos de ejemplo: meses y ventas
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [150, 200, 180, 250, 300]

# Crear el gráfico de barras
plt.bar(meses, ventas, color='red')

# Agregar etiquetas y título
plt.xlabel('Meses')
plt.ylabel('Ventas (en miles)')
plt.title('Ventas Mensuales')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo: meses y ventas
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
ventas = [150, 200, 180, 250, 300]

# Crear el diagrama de torta
plt.pie(ventas, labels=meses, autopct='%1.1f%%', startangle=90, colors=['red', 'orange', 'yellow', 'green', 'blue'])

# autopct='%1.1f%%': Esto agrega etiquetas con el porcentaje de cada porción en el diagrama.

# Agregar título
plt.title('Distribución de Ventas Mensuales')

# Mostrar el diagrama de torta
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo: horas de estudio y calificaciones
horas_estudio = [2, 3, 4, 5, 6, 7, 8]
calificaciones = [60, 65, 70, 75, 80, 85, 90]

# Crear el gráfico de dispersión
plt.scatter(horas_estudio, calificaciones, color='green', marker='o')

# Agregar etiquetas y título
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre Horas de Estudio y Calificaciones')

# Mostrar el gráfico
plt.show()
```

```python
file = 'carros.xlsx'
data = pd.read_excel(file)
df = pd.DataFrame(data)
display(df)

#   name: Nombre
#   year: Año
#   selling_price: Precio de venta
#   km_driven: Kilómetros recorridos
#   fuel: Combustible
#   seller_type: Tipo de vendedor
#   transmission: Transmisión
#   owner: Propietario
```

```python
plt.scatter(x=[1,2,3], y=[5,3,10], color = 'red')
```

```python
plt.scatter(x=df.selling_price, y=df.km_driven, color='red', marker='+')
```

```python
df_sorted = df.sort_values(by='selling_price', ascending=False)

# Toma los 10 productos con los precios de venta más altos
top_10_products = df_sorted.head(10)

# Crea la gráfica de dispersión
plt.figure(figsize=(10, 6))  # Ajusta el tamaño de la figura según tus preferencias
plt.scatter(x=top_10_products.selling_price, y=top_10_products.name, color='red', marker='+')

# Personaliza la gráfica (agrega etiquetas, título, etc.)
plt.xlabel('Precio de Venta')
plt.ylabel('Nombre del Producto')
plt.title('Top 10 Productos con Precios de Venta Más Altos')

# Muestra la gráfica
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10_products['selling_price'], labels=top_10_products['name'], autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Productos con Precios de Venta Más Altos')
plt.axis('equal')  # Para que el círculo sea perfecto
plt.show()
```
